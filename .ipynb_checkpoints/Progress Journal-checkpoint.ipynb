### Introduction

#### a.	Reason

Electricity production or consumption in Turkey acts as a kind of stock market. Therefore, hourly forecasting is essential for manufacturers and consumers. In addition, orders can be assigned with these forecasts during the day.
Here, consumers want to avoid the obligation to pay more, which is determined by a coefficient above the specified pricing, while producers want to avoid a decrease in prices.
Of course, it is difficult to make an ideal forecast due to various reasons such as weather conditions, public holidays, or the volatility of the exchange rate. Nevertheless, despite all these negativities, the teams responsible for this work of producer or consumer companies have a duty to minimize the margin of error.
Positive imbalance: It is interpreted as producer firms producing more than expected or consumer firms consuming less than expected.
Negative imbalance: It is interpreted as producer firms producing less than expected or consumer firms consuming more than expected.
Thus, the net imbalance data to be estimated actually emerges as the difference of these two imbalances.
According to the estimation of this net balance sign, in intraday orders, consumer companies see whether they will pay a penalty price when placing orders, while producer companies see the price of the electricity they will sell.
Identifying orders early is just as important as getting the estimation right.

#### b.	Aim of the study

As a producer or consumer firm, estimating the net imbalance sign at hour t at the beginning of hour t-12. (Actually just before the 12 pm we can predict the net imbalance of the next 12 hours)

#### c.	Baseline

We choose linear regression models with all parameters and improve that baseline with use some specific parameters according to their interception values.
#### d.	Approach

Considering the changing needs and uncertain economic situation of the country, it has been determined that it is more reasonable to evaluate this daily basis while making intraday forecasts for t hour. Actually, we can predict 12 pm-12 am time interval just before 12 pm.
As you can see on our code the most reasonable parameters are selected depends on their interception values and the results are getting better.
#### e. Conclusion

As a result of the calculations made with the importance given to parameters, an improvement is seen compared to the baseline methods.
Considering the adequacy of this improvement as another matter of discussion, a more efficient estimation is of course possible.
Evaluation was made directly according to trial time data.
